# t-SNE Corpus Visualization

We will try to visualize our corpus with t-SNE.

```python
from yellowbrick.text import TSNEVisualizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
merged = pd.read_csv('../../DataPlus/feature_dataframe.csv')
```

```python
df = merged.dropna(subset=['Convo_1'])
tfidf = TfidfVectorizer()
docs = tfidf.fit_transform(df['Convo_1'])
```

```python
def tsne(df, var):
    tsne = TSNEVisualizer()
    
    # drops examples with NA for var of interest
    drop_df = df.dropna(subset=[var])
    labels = drop_df[var]
    
    # fit and visualize
    tsne.fit(docs, labels)
    tsne.poof()
```

### Label: txgot_binary

Treatment patient received (1: active surveillance, 0: no AS)

```python
tsne(df, 'txgot_binary')
```

### Label: Dataset

Origin dataset of patient

```python
tsne(df, 'Dataset')
```

```python
tsne(df, 'Doctor_1')
```
